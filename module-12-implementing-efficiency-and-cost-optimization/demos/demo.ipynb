```python
import pandas as pd
```

```python
#Load baseline and post-optimization metrics from CSVs

BASELINE_PATH = "baseline_genai_metrics.csv"
POST_PATH = "post_optimized_genai_metrics.csv"

baseline_df = pd.read_csv(BASELINE_PATH)
post_df = pd.read_csv(POST_PATH)
```

```python
post_df
```

```python
#Build a simple before/after comparison table

baseline = baseline_df.iloc[0]
post = post_df.iloc[0]

metrics = [
    ("p50_latency_ms", "P50 latency (ms)"),
    ("p95_latency_ms", "P95 latency (ms)"),
    ("tokens_per_second", "Throughput (tokens/sec)"),
    ("avg_power_watts", "Avg power (watts)"),
    ("cost_per_1k_tokens_usd", "Cost per 1k tokens ($)"),
    ("quality_score", "Quality score (0–1)"),
]

rows = []
for key, label in metrics:
    b = float(baseline[key])
    a = float(post[key])
    change_pct = ((a - b) / b * 100) if b != 0 else None
    rows.append({
        "metric": label,
        "baseline": b,
        "post_optimization": a,
        "change_%": "n/a" if change_pct is None else f"{change_pct:+.1f}%",
    })

comparison_df = pd.DataFrame(rows)
comparison_df
```

```python
print("Baseline vs Post-Optimization (simulated)")
print(f"P95 latency: {int(baseline['p95_latency_ms'])} ms -> {int(post['p95_latency_ms'])} ms")
print(f"Throughput: {baseline['tokens_per_second']} -> {post['tokens_per_second']} tokens/sec")
print(f"Avg power: {baseline['avg_power_watts']} -> {post['avg_power_watts']} watts")
print(f"Cost per 1k tokens: ${baseline['cost_per_1k_tokens_usd']} -> ${post['cost_per_1k_tokens_usd']}")
print(f"Quality score: {baseline['quality_score']} -> {post['quality_score']}")

print("\nDecision prompt:")
print("Is the reduction in cost and power worth the increase in tail latency and the small quality drop for this workload?")
```
