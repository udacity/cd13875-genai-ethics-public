```python
import mermaid as md
```

```python
md.Mermaid("""
flowchart LR

%% Demo: Human Review Workflow with Risk-Based Routing

A["User requests content"]
B["GenAI generates draft"]

%% Risk classification decision
C{"High-risk content? \n(financial, legal, safety, hiring)"}

%% Automated baseline path
F["Draft sent"]

%% Human review path
D["Human review step"]
E{"Approved?"}
G["Revise or escalate"]

%% Core generation flow
A --> B --> C

%% Low-risk = automatic publish
C -- "No" --> F

%% High-risk = mandatory human review
C -- "Yes" --> D
D --> E
E -- "Yes" --> F
E -- "No" --> G --> B

""")
```
